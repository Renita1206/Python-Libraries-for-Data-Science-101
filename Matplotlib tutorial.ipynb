```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot([1,3,2,5],[2,6,4,10])
plt.title("Test Plot", color="w")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.figure(figsize=(2,5))
plt.plot([1,3,2,5],[2,6,4,10],"ro") #ro for red dot, kx for black cross
plt.axis([0,6,0,20]) # axis limits => x is 0 to 6 and y is 0 to 20
plt.annotate("smh",(4,1))
plt.show()
```

```python
plt.clf() #clear figure
x=np.arange(4)
y=[3.3,5.6,7,9]
plt.xticks(x,("A","B","C","D"))
plt.bar(x,y, color=["r","k","b","y"])
plt.show()
```

```python
d={
    "r":[1,2,3,4],
    "g":[2,4,6,8]
}
df=pd.DataFrame(d)
df.plot(style=["rx","gx"])
#plt.savefig("name.png") to save figure
plt.show()
```

```python
fig = plt.figure()
fig.suptitle('My SubPlots')
fig.add_subplot(221)
plt.plot([np.log(n) for n in range(1,10)])
fig.add_subplot(222, facecolor='y')
fig.add_subplot(223)
fig.add_subplot(224) 
plt.show()
```

```python
fig, plots = plt.subplots(2, sharex=True)
fig.suptitle('Sharing X axis')
x = range(0,200,5)
y = [n**0.8 for n in x]
plots[0].plot(x, y, color='r')
plots[1].scatter(x, y)
plt.show()
```

```python
#time series
ts = pd.Series(np.random.randn(180), index=pd.date_range('1/1/2018', periods=180))
df = pd.DataFrame(np.random.randn(180, 3), index=ts.index, columns=list('ABC'))
df.cumsum().plot() #cumsum for cumulative sum
plt.show()
```

```python
#scatter
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
sizes = (30 * np.random.rand(N))**2   # 0 to 15 point radii
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5)
plt.show()
```
